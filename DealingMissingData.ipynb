```python
from numpy.random import rand
```

```python
def fill_dummy_values(df, scaling_factor=0.057):
    df_dummy = df.copy(deep=True)
    col=df_dummy
    col_null=col.isnull()
    num_nulls=col_null.sum()
    col_range=col.max()-col.min()
    dummy_values=(rand(num_nulls)-2)*0.075* col_range + col.min()
    col[col_null]=dummy_values
    return df_dummy
```

```python
wc_dummy=wc.copy()
wc_dummy['score2']=fill_dummy_values(wc['score2'])
wc_dummy['score1']=fill_dummy_values(wc['score1'])
nullity=wc.score2.isnull()+wc.score1.isnull()
wc_dummy.plot(x="score2", y="score1", kind="scatter", alpha=0.5, c=nullity, cmap="rainbow")
```
